```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/test.csv')
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/train.csv')

# Check data
print(df.head())
```

```python
# Separate features and target
X = df.drop('price_range', axis=1)  # Features (all columns except target)
y = df['price_range']   # Target variable

# Standardize the features
scaler = StandardScaler() #perform better on similar scales
X_scaled = scaler.fit_transform(X)

# Split into train and test
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
print(f"Training set size: {X_train.shape[0]} samples")
print(f"Test set size: {X_test.shape[0]} samples")
```

```python
# Initialize the model using iterations for backpropagation
mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, alpha=0.0001,
                    solver='adam', random_state=42, verbose=True)

# Train the model
mlp.fit(X_train, y_train)
```

```python
# Predict
y_pred = mlp.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Detailed metrics
print("Classification Report:\n")
print(classification_report(y_test, y_pred))
```
